```python
def extract_minutes(duration):
    parts = duration.split(' ')
    minutes = 0
    for part in parts:
        if 'h' in part:
            minutes += int(part.replace('h', '')) * 60
        elif 'm' in part:
            minutes += int(part.replace('m', ''))
    return minutes

final_df['duration'] = final_df['Duration'].apply(extract_minutes)
```
